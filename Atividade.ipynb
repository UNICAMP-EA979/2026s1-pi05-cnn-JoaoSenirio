### Data Module

A classe [pytorch_lightning.LightningDataModule](https://lightning.ai/docs/pytorch/stable/data/datamodule.html) pode ser utilizada para organizar os DataLoaders e os dados.

```python
class MNISTDataModule(pl.LightningDataModule):
    def __init__(self, hparams):
        super().__init__()
        self.save_hyperparameters(hparams)

    def setup(self, stage=None):
        self.train = training_data
        self.val = val_data

    '''
    Dataloaders optimize data loading and automatically prepare batches
    '''
    def train_dataloader(self):
        return DataLoader(self.train,
                            batch_size=self.hparams.batch_size,
                            num_workers=self.hparams.nworkers,
                            shuffle=True,
                            persistent_workers=True,
                            pin_memory=True)

    def val_dataloader(self):
        return DataLoader(self.val,
                          batch_size=self.hparams.batch_size,
                          num_workers=self.hparams.nworkers,
                          shuffle=False,
                          persistent_workers=True,
                          pin_memory=True)
```

### Exercício 1.2

Para que serve cada camada da rede? Por que foi necessário modificar ligeiramente a rede (célula anterior) e que modificações foram realizadas?

RESPOSTA:

Inicialmente, a imagem passa por uma camada de convolução (conv1), que o papel é extrair características básicas das imagens de entrada. Em seguida, a saída passa por uma normalização (bn1), e por uma função de ativação ReLU (relu), que introduz não linearidade. Depois, é aplicado um pooling (maxpool), que reduz a dimensão espacial da imagem, capturando as informações mais relevantes.

Após isso, vem 4 camadas, de forma que todas são compostas, por uma convolução, normalização, ReLu, uma segunda convolução e uma segunda normalização

Na layer1, a rede trabalha com 64 features-map, extraindo padrões simples (bordas, linhas horizontais/verticais por exemplo). Na layer2, ocorre um aumento no número de features para 128, por meio de um downsampling. Nesse ponto, a rede começa a aprender padrões mais complexos. Na layer3, há novo downsampling e o número de canais sobe para 256. Por fim, na layer4 há o último aumento de features, chegando a 512. Nesse ponto, a rede já extrai representações bem abstratas, úteis para distinguir as classes.

Finalmente, a saída passa pela avgpool (Adaptive Average Pooling), que compacta cada canal em um único valor de média. Por fim, esse vetor é enviado para a camada totalmente conectada (fc), responsável por produzir os escores finais e decidir qual é a classe mais provável da imagem.

Foi necessário modificar ligeiramente a rede porque no presente caso  o conjunto possui imagens em escala de cinza, isto é, apenas 1 canal, bem como o objetivo final é de classificar dígitos em 10 classes (de 0 a 9).
Dessa forma, a primeira camada convolucional foi adaptada para receber 1 canal de entrada em vez de 3, bem como a camada final foi alterada para produzir 10 saídas em vez de 1000.

### Exercício 1.3

Tente entender as dimensões da saída de cada camada da rede. Por exemplo, porque as três primeiras camadas possuem saídas com dimensões (-1, 64, 14, 14) e a quarta camada tem saída com dimensões (-1, 64, 7, 7)?

E quanto aos parâmetros a serem aprendidos, você consegue explicar como eles são calculados?

RESPOSTA:

O formato de saída de cada camada pode ser escrito como (a, b, c, d), em que temos: batch, número de feature maps, altura e largura. No caso da entrada da rede, a imagem possui shape [-1, 1, 28, 28], isto é, um batch de tamanho variável, 1 canal e resolução espacial de 28x28. O valor -1 indica que o tamanho do batch não está fixado.

Ao passar pela primeira convolução, a entrada é transformada em [-1, 64, 14, 14]. Isso acontece porque essa camada aplica 64 filtros, gerando assim 64 feature maps de saída. Além disso, a dimensão espacial é reduzida de 28x28 para 14x14 em razão do stride = 2, que define o passo com que o filtro percorre a imagem. Em seguida, essas feature maps passam por uma camada de MaxPool2d, também com stride = 2, reduzindo novamente a dimensão espacial, de modo que a saída passa a ser [-1, 64, 7, 7].

A partir desse ponto, nas 4 camadas seguintes o número de feature maps aumenta, enquanto a dimensão espacial vai sendo reduzida. Na layer1, contudo, isso ainda não acontece: o shape permanece [-1, 64, 7, 7]. Já na layer2, o primeiro bloco realiza downsampling. A primeira convolução desse bloco transforma a entrada em uma saída com shape [-1, 128, 4, 4], pois o número de canais passa de 64 para 128 e a dimensão espacial é reduzida devido ao uso de stride = 2. As demais operações do bloco não alteram esse formato. Embora exista uma segunda convolução nesse mesmo bloco, ela utiliza stride = 1, e por isso mantém as dimensões espaciais. O mesmo padrão se repete nas camadas seguintes: a layer3 gera saída [-1, 256, 2, 2], e a layer4 gera saída [-1, 512, 1, 1], sempre porque a primeira convolução do primeiro bloco dessas camadas realiza downsampling e aumenta o número de canais.

Com relação aos parâmetros aprendidos, nas camadas de convolução eles dependem do número de features_maps de entrada, de saída e do tamanho do kernel. A fórmula geral é: in_channels x out_channels x kernel_height x kernel_width. No caso da primeira convolução da rede, temos 64 canais de saída, 1 canal de entrada e kernel 7x7. Assim, o número de parâmetros é 64 x 1 x 7 x 7 = 3136.

Na normalização, mais especificamente na BatchNorm2d, são aprendidos 2 parâmetros por feature map: um parâmetro de escala e um de deslocamento. Portanto, na primeira normalização, como há 64 feature maps, o número de parâmetros aprendidos é 2 x 64 = 128.

Por fim, operações como ReLU e Max Pooling não possuem parâmetros a serem aprendidos.

### Lightning Module

Aqui definimos o comportamento durante o treinamento, ou seja, o que acontece com um lote retornado pelo DataLoader.

```python
class TrainMNIST(pl.LightningModule):
    def __init__(self, hparams):
        super().__init__()

        self.save_hyperparameters(hparams)
        self.model = cnn_model
        self.loss = nn.CrossEntropyLoss()
        self.accuracy = Accuracy(task="multiclass", num_classes=self.hparams.nout)

        torchinfo.summary(self.model)

    def forward(self, x):
        return self.model(x).softmax(dim=-1)

    def classification_step(self, mode, batch):
        x, y = batch
        y_hat = self.forward(x)
        loss = self.loss(y_hat, y)

        if mode == "train":
            self.log("loss", loss, on_epoch=True, on_step=True)
            return loss
        elif mode == "val":
            _, preds = torch.max(y_hat, 1)
            self.accuracy(preds, y)
            self.log("val_loss", loss, on_epoch=True, on_step=False, prog_bar=True)
            self.log("val_acc", self.accuracy, on_epoch=True, on_step=False, prog_bar=True)

    def training_step(self, train_batch, batch_idx):
        return self.classification_step("train", train_batch)

    def validation_step(self, val_batch, batch_idx):
        self.classification_step("val", val_batch)

    def configure_optimizers(self):
        optimizer = AdamW(self.model.parameters(), lr=self.hparams.lr, weight_decay=self.hparams.wd)

        return optimizer
```

### Treinando a CNN

```python
debug = False  # Turn this True to test the code with a quick run

# Initialization
model = TrainMNIST(hparams)
data = MNISTDataModule(hparams)

# The PL trainer can run various callbacks during training. Here we only use the
# callback that saves the weights of the trained network.
os.makedirs("logs", exist_ok=True)
checkpoint_callback = [ModelCheckpoint(dirpath="logs",  # path to save checkpoint
                                       filename=hparams["experiment_name"] + "-{epoch}-{val_acc:.2f}",
                                       monitor="val_acc",
                                       mode="max")]

trainer = pl.Trainer(max_epochs=hparams["max_epochs"], #devices=1, accelerator="gpu",
                     fast_dev_run=debug,
                     callbacks=checkpoint_callback,
                     log_every_n_steps=1
                     )

print("Hyperparameters:\n")
for k, v in hparams.items():
    print(f'{k}: {v}')

trainer.fit(model, data)
```

### Análise dos resultados

Após o treinamento da rede, é importante checar os resultados de validação, visualizando algumas saídas da rede e obtendo métricas de classificação. Abaixo exemplificamos a visualização da matriz de confusão.

```python
# WARNING: This will load a random checkpoint from the logs folder, if you want an specific one change it!
baseline_model = random.choice(glob("logs/*.ckpt"))
print(f"Using model {baseline_model}")

# Load trained model
model = TrainMNIST.load_from_checkpoint(baseline_model).eval().cuda()

# Initialize validation data
data = MNISTDataModule(hparams)
data.setup()
data.hparams.batch_size = 1
val_data = data.val_dataloader()

def collect_2d_results(data, model):
    y_trues = []
    y_labels = []
    for (img, label) in data:
        with torch.no_grad():
            out = model(img.cuda()).argmax(dim=-1).item()

        label = label.numpy()
        y_trues.append(label)
        y_labels.append(out)

    return y_trues, y_labels

y_trues_val, y_labels_val = collect_2d_results(val_data, model)
```
